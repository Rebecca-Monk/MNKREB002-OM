Calculate the Taylor Series expansion for the following univariate functions f(x) about the point x = x0. Only show the series terms until the truncation error is o(x − x0)^3 and upload a picture or a document on your GitHub:
#### Functions

#### Derivatives
1. $f'(x)=\frac{1}{(x+1)^2}$
2. $f'(x)=\frac{-4}{x^3}$

#### Taylor Series Expansions about $x = x_0$

we use:
$f(x)=f(x_0)+f'(x_0)(x-x_0)+\frac{f''(x_0)}{2}(x-x_0)^2+o\left((x-x_0)^3\right)$

#### 1. $f(x)=\frac{x}{x+1}$

derivatives:
$$ f'(x)=\frac{1}{(x+1)^2}, \quad f''(x)=-\frac{2}{(x+1)^3} $$

at $x_0$:
$$ f(x_0)=\frac{x_0}{x_0+1}, \quad f'(x_0)=\frac{1}{(x_0+1)^2}, \quad f''(x_0)=-\frac{2}{(x_0+1)^3} $$

Taylor Expansion:
$$ \frac{x}{x+1} = \frac{x_0}{x_0+1}
+
\frac{1}{(x_0+1)^2}(x-x_0)
-
\frac{1}{(x_0+1)^3}(x-x_0)^2
+
o\left((x-x_0)^3\right)
$$

---

#### 2. $f(x)=\frac{2x^2}{x^4} = \frac{2}{x^2}$

derivatives:
$$
f'(x)=-\frac{4}{x^3}, \quad
f''(x)=\frac{12}{x^4}
$$

at $x_0$:
$$
f(x_0)=\frac{2}{x_0^2}, \quad
f'(x_0)=-\frac{4}{x_0^3}, \quad
f''(x_0)=\frac{12}{x_0^4}
$$

Taylor Expansion:
$$
\frac{2}{x^2}
=
\frac{2}{x_0^2}
-
\frac{4}{x_0^3}(x-x_0)
+
\frac{6}{x_0^4}(x-x_0)^2
+
o\left((x-x_0)^3\right)
$$

---

#### 3. $f(x)=\ln(x^2)$

derivatives:
$$
f'(x)=\frac{2}{x}, \quad
f''(x)=-\frac{2}{x^2}
$$

at $x_0$ (with $x_0 \neq 0$):
$$
f(x_0)=\ln(x_0^2), \quad
f'(x_0)=\frac{2}{x_0}, \quad
f''(x_0)=-\frac{2}{x_0^2}
$$

Taylor Expansion:
$$
\ln(x^2)
=
\ln(x_0^2)
+
\frac{2}{x_0}(x-x_0)
-
\frac{1}{x_0^2}(x-x_0)^2
+
o\left((x-x_0)^3\right)
$$

```python
# Taylor Series Expansion Calculator
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function 1: f(x) = x/(x+1)
print("\n1. f(x) = x/(x+1) about x₀ = 0")
print("-"*80)

def f1(x):
    return x / (x + 1)

def f1_prime(x):
    return 1 / (x + 1)**2

def f1_double_prime(x):
    return -2 / (x + 1)**3

def f1_triple_prime(x):
    return 6 / (x + 1)**4

# at x0 = 0
x0_1 = 0
f1_val = f1(x0_1)                     # f(0) = 0
f1_prime_val = f1_prime(x0_1)         # f'(0) = 1
f1_double_val = f1_double_prime(x0_1) # f''(0) = -2
f1_triple_val = f1_triple_prime(x0_1) # f'''(0) = 6

print(f"f({x0_1}) = {f1_val}")
print(f"f'({x0_1}) = {f1_prime_val}")
print(f"f''({x0_1}) = {f1_double_val}")
print(f"f'''({x0_1}) = {f1_triple_val}")
print(f"\nTaylor Series: f(x) ≈ 0 + 1·x + (-2/2!)·x² + (6/3!)·x³")
print(f"              f(x) ≈ x - x² + x³ + O(x⁴)")

# Taylor Series Function
def taylor1(x):
    return x - x**2 + x**3
```

```python
# Function 2: f(x) = 2x²/x⁴ = 2/x²
print("\n\n2. f(x) = (2x²)/(x⁴) about x₀ = 1")
print("-"*80)

def f2(x):
    return (2*x**2) / (x**4)

f2_val = 2
f2_prime = -4
f2_double = 12

print(f"f(1) = {f2_val}")
print(f"f'(1) = {f2_prime}")
print(f"f''(1) = {f2_double}")
print(f"\nTaylor Series: f(x) ≈ 2 - 4(x-1) + 6(x-1)² + O((x-1)³)")

def taylor2(x):
    h = x - 1
    return 2 - 4*h + 6*h**2
```

```python
# Function 3: f(x) = ln(x²)
print("\n\n3. f(x) = ln(x²) about x₀ = 1")
print("-"*80)

def f3(x):
    return np.log(x**2)

def f3_prime(x):
    return 2/x

def f3_double_prime(x):
    return -2/x**2

def f3_triple_prime(x):
    return 4/x**3

# at x0 = 1
x0_3 = 1
f3_val = f3(x0_3)                     # f(1) = 0
f3_prime_val = f3_prime(x0_3)         # f'(1) = 2
f3_double_val = f3_double_prime(x0_3) # f''(1) = -2
f3_triple_val = f3_triple_prime(x0_3) # f'''(1) = 4

print(f"f({x0_3}) = {f3_val}")
print(f"f'({x0_3}) = {f3_prime_val}")
print(f"f''({x0_3}) = {f3_double_val}")
print(f"f'''({x0_3}) = {f3_triple_val}")

print(f"\nTaylor Series: f(x) ≈ 0 + 2·(x-1) + (-2/2!)·(x-1)² + (4/3!)·(x-1)³")
print(f"              f(x) ≈ 2(x-1) - (x-1)² + (2/3)(x-1)³ + O((x-1)⁴)")

# Taylor Series Function
def taylor3(x):
    h = x - 1
    return 2*h - h**2 + (2/3)*h**3
```

```python
# VISUALIZATIONS
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# Function 1
x1 = np.linspace(-0.8, 0.8, 300)
x1 = x1[x1 != -1]
f1_vals = f1(x1)
taylor1_vals = taylor1(x1)

axes[0].plot(x1, f1_vals, 'b-', linewidth=2.5, label='f(x) = x/(x+1)')
axes[0].plot(x1, taylor1_vals, 'r--', linewidth=2.5, label='Taylor Series')
axes[0].axvline(x=0, color='gray', linestyle=':', alpha=0.5)
axes[0].scatter([0], [0], color='green', s=100, zorder=5)
axes[0].set_xlabel('x', fontsize=11)
axes[0].set_ylabel('f(x)', fontsize=11)
axes[0].set_title('1. f(x) = x/(x+1) about x₀ = 0', fontsize=12, fontweight='bold')
axes[0].legend(fontsize=10)
axes[0].grid(True, alpha=0.3)
axes[0].set_xlim([-0.8, 0.8])
axes[0].set_ylim([-1, 2])

# Function 2
x2 = np.linspace(0.5, 1.5, 300)
axes[1].plot(x2, f2(x2), 'b-', linewidth=2.5, label='f(x) = (2x²)/(x⁴)')
axes[1].plot(x2, taylor2(x2), 'r--', linewidth=2.5, label='Taylor Series')
axes[1].axvline(x=1, color='gray', linestyle=':', alpha=0.5)
axes[1].scatter([1], [2], color='green', s=100, zorder=5)
axes[1].set_xlabel('x', fontsize=11)
axes[1].set_ylabel('f(x)', fontsize=11)
axes[1].set_title('2. f(x) = (2x²)/(x⁴) about x₀ = 1', fontsize=12, fontweight='bold')
axes[1].legend(fontsize=10)
axes[1].grid(True, alpha=0.3)
axes[1].set_ylim([0, 5])

# Function 3
x3 = np.linspace(0.1, 2, 300)
f3_vals = f3(x3)
taylor3_vals = taylor3(x3)

axes[2].plot(x3, f3_vals, 'b-', linewidth=2.5, label='f(x) = ln(x²)')
axes[2].plot(x3, taylor3_vals, 'r--', linewidth=2.5, label='Taylor Series')
axes[2].axvline(x=1, color='gray', linestyle=':', alpha=0.5)
axes[2].scatter([1], [0], color='green', s=100, zorder=5)
axes[2].set_xlabel('x', fontsize=11)
axes[2].set_ylabel('f(x)', fontsize=11)
axes[2].set_title('3. f(x) = ln(x²) about x₀ = 1', fontsize=12, fontweight='bold')
axes[2].legend(fontsize=10)
axes[2].grid(True, alpha=0.3)
axes[2].set_xlim([0.1, 2])

plt.suptitle('Taylor Series Expansions (Truncation Error: O((x-x₀)³))', fontsize=14, fontweight='bold', y=1.00)
plt.tight_layout()
plt.savefig('taylor_series_results.png', dpi=300, bbox_inches='tight')
plt.show()

print("Plot saved as 'taylor_series_results.png'")
```

Consider the implicit Euler backward scheme:
$$C_{n+1} = \frac{C_n}{1 + k\Delta t}$$
- This scheme is unconditionally stable
- Demonstrate that this is true by using the same method used in the previous slide
- Upload a picture or a document

```python
#%% IMPORTS
import numpy as np
import matplotlib.pyplot as plt

#%% PARAMETERS
k = 0.5                 # decay constant
C0 = 100                # initial concentration
t_end = 20              # total simulation time

# Timesteps to compare
dt_values = [0.1, 1, 3, 5]

#%% FIGURE SETUP
fig, ax = plt.subplots(1, 3, figsize=(18,5))

# EXPLICIT EULER
for dt in dt_values:

    time = np.arange(0, t_end + dt, dt)
    C_exp = np.zeros(len(time))
    C_exp[0] = C0

    for n in range(len(time)-1):
        C_exp[n+1] = (1 - k*dt) * C_exp[n]

    ax[0].plot(time, C_exp, marker='o', label=f'Δt={dt}')

ax[0].set_title('Explicit Euler Scheme')
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Concentration')
ax[0].set_ylim(-600, 600)
ax[0].grid(True)
ax[0].legend()

# IMPLICIT EULER
for dt in dt_values:

    time = np.arange(0, t_end + dt, dt)
    C_imp = np.zeros(len(time))
    C_imp[0] = C0

    for n in range(len(time)-1):
        C_imp[n+1] = C_imp[n] / (1 + k*dt)

    ax[1].plot(time, C_imp, marker='o', label=f'Δt={dt}')

ax[1].set_title('Implicit Euler Scheme')
ax[1].set_xlabel('Time')
ax[1].set_ylabel('Concentration')
ax[1].grid(True)
ax[1].legend()

# AMPLIFICATION FACTORS
dt_range = np.linspace(0, 6, 200)

G_explicit = np.abs(1 - k*dt_range)
G_implicit = np.abs(1 / (1 + k*dt_range))

ax[2].plot(dt_range, G_explicit, label='Explicit')
ax[2].plot(dt_range, G_implicit, label='Implicit')

ax[2].axhline(1, linestyle='--', label="kt = 1")

ax[2].set_title('Amplification Factor')
ax[2].set_xlabel('Time')
ax[2].set_ylabel('|G|')
ax[2].grid(True)
ax[2].legend()

#%% FORMAT
plt.tight_layout()
plt.show()
plt.savefig('explicit_implicit_euler.png', dpi=300, bbox_inches='tight')
```

## Ocean Modelling: Exercise 6
- Modify the Euler-forward method implemented in the codes at the beginning of this section (either the python or matlab version) and change it into the Euler-backward method
- Compare the results for different values of ∆t and upload your code and results on the GitHub repository

```python
# Compare Forward and Backward Euler Methods
import pylab as pl

# parameters
t0 = 0.0
tn = 20.0
kappa = 0.05
C0 = 0.5

# timesteps to test
dt_values = [0.1, 1, 5, 10]

# create figure
fig, ax = pl.subplots(1, 2, figsize=(14,6))

for dt in dt_values:

    # number of timesteps
    Ntot = int(pl.floor((tn - t0)/dt) + 1)
    # arrays
    t = pl.zeros(Ntot)

    C_forward = pl.zeros(Ntot)
    C_backward = pl.zeros(Ntot)

    # initial conditions
    t[0] = t0
    C_forward[0] = C0
    C_backward[0] = C0

    # time stepping loop
    for n in range(Ntot - 1):

        # Forward Euler (explicit)
        C_forward[n+1] = C_forward[n] * (1 - kappa*dt)

        # Backward Euler (implicit)
        C_backward[n+1] = C_backward[n] / (1 + kappa*dt)

        # update time
        t[n+1] = t[n] + dt

    # plot
    ax[0].plot(t, C_forward, marker='o', label=f'Δt={dt}')
    ax[1].plot(t, C_backward, marker='o', label=f'Δt={dt}')

# formatting Forward Euler plot
ax[0].set_title("Forward Euler (Explicit)")
ax[0].set_xlabel("Time (days)")
ax[0].set_ylabel("Concentration")
ax[0].grid(True)
ax[0].legend()

# formatting Backward Euler plot
ax[1].set_title("Backward Euler (Implicit)")
ax[1].set_xlabel("Time (days)")
ax[1].set_ylabel("Concentration")
ax[1].grid(True)
ax[1].legend()

pl.tight_layout()
pl.show()
plt.savefig('forward_backward_euler.png', dpi=300, bbox_inches='tight')
```
