#### Questions
#### 1. Is the growth exponential?
There is exponential growth from day 5 which is seen when plotting cell number, biovolume, cell carbon and chloropyll-a over time (seen in Figure 1). The exponential growth stops after various days depending on the variable used to determine the cell growth. There is no exponential growth in the ammonium plot (there is a decay over time).

#### 2. Can you separate the growth curve into some phases?
When looking at the cell number over time, 3 distinct phases can be identified. These are: the lag phase which occurs from day 0 to day 1, the exponential growth phase which occurs between day 5 and day 18, the stationary phase which occurs between day 18 and day 25. These phases are indicates on Figure 2. 

#### 3. How different is the growth expressed in cellular carbon from the one in cell number and chlorophyll?
The 3 growth curves are quite different. For cell carbon, the exponential growth starts the earliest and continues to increase until day 25. For chlorophyll-a, the exponential growth also starts quite early, but later than the cell carbon. The exponential growth phase also ends earlier for chlorophyll-a and instead of there being a stationary phase, the chlorophyll-a decreases after reaching a maximum at day 16. The cell count's exponential growth starts at a similar day to chlorophyll-a and reaches it's maximum later at day 18. After this, the stationary phase is reached and the cell count remains at this maximum.

#### 4. Is the ammonium consumption approximated by an exponential curve?
The ammonium consumption is exponential (which can be seen when inverting the y-axis as done in Figure 3 as it shows the typical shape of an exponential growth curve). However, the ammonium consumptions follows an exponential decay rather an exponential growth. Once all of the ammonium is consumed, the growth seen in the other curves also reaches the stationary phase after a slight lag.

#### 5. How would you estimate the parameters of the growth curve?
the formula for exponential growth is: $$N_t = N_0 \cdot e^{\mu t}$$
the parameters are: N0 (initial cell number), µ (growth rate per day) and t (doubling time)
- N0 will be calculated at day 5 where the exponential growth phase starts
- µ can be calculated as the gradient of the straight-line graph that models the exponential growth phase
- t is calculated using the following formula: $$t_{doubling} = \frac{\ln(2)}{\mu}$$



```python
# Growth Phases for Cell Number vs Time 
# vertical lines to distinguish growth phases
phases = [
    {'x': 5,  'label': 'Start of Exponential Growth (D5)', 'colour': 'green', 'linestyle': '--'},
    {'x': 18,  'label': 'End of Exponential Growth (D18)', 'colour': 'red', 'linestyle': '--'},]
# ploting 
fig, ax = plt.subplots(figsize=(12, 6))
# cell count line
ax.plot(df['time'], df['cells'], color='blue', linewidth=2, marker='o', markersize=4, label='Cells')
# vertical phase lines
for phase in phases:
    ax.axvline(x=phase['x'], color=phase['colour'], linestyle=phase['linestyle'], linewidth=1.5, label=phase['label'])
# labels  and formatting
ax.set_title('Figure 2: Phytoplankton Cell Growth Phases Over Time', fontsize=16, fontweight='bold')
ax.set_xlabel('Time (days)', fontsize=13)
ax.set_ylabel('Cell Count (cells/mL)', fontsize=13)
ax.legend(fontsize=11)
ax.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

Write a python notebook to plot the analytical solution of the model presented in Eq. (4) to show the time evolution of phytoplankton and nutrient concentrations. Use the following parameters:
- time axis in days from 0 to 20 days 
- k′ = 0.1 d−1
- P0 = 0.1 mmol N m−3 
- N0 = 5 mmol N m−3

Equation 4: $$P(t) = \frac{(P_0 + N_0) P_0}{P_0 + N_0 e^{-k'(P_0 + N_0)t}}$$
this equation describes the logistic growth model where the growth of the consumer is limited by the amount of resources available - the growth of phytoplankton is limited by the nutrient concentrations

```python
# Plotting the Analytical Solution
# parameters 
k_prime = 0.1    # d⁻¹
P0      = 0.1    # mmol N m⁻³
N0      = 5.0    # mmol N m⁻³
total_PN = P0 + N0  # total nutrients (conserved quantity)
# time axis 
t = np.linspace(0, 20, 500)  # 0 to 20 days
# analytical solutions 
P = ((P0 + N0) * P0) / (P0 + N0 * np.exp(-k_prime * (P0 + N0) * t))
N = N0 + P0 - P   # nutrient conservation
# subplots 
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('Figure 4: Logistic Growth: Analytical Solution', fontsize=18, fontweight='bold')

# 1. Phytoplankton Growth 
ax1.plot(t, P, color='blue', linewidth=2.5, label='P(t)')
ax1.axhline(y=total_PN, color='black', linewidth=1.5, linestyle='--', label=f'P₀ + N₀ = {total_PN}')
ax1.set_title('Phytoplankton Biomass', fontsize=13, fontweight='bold')
ax1.set_xlabel('Time (days)', fontsize=11)
ax1.set_ylabel('Biomass (mmol N m⁻³)', fontsize=11)
ax1.legend(fontsize=10)
ax1.grid(True, linestyle='--', alpha=0.4)

# 2. Nutrient Decay 
ax2.plot(t, N, color='green', linewidth=2.5, label='N(t)')
ax2.axhline(y=total_PN, color='black', linewidth=1.5, linestyle='--', label=f'P₀ + N₀ = {total_PN}')
ax2.set_title('Nutrient Concentration', fontsize=13, fontweight='bold')
ax2.set_xlabel('Time (days)', fontsize=11)
ax2.set_ylabel('Nutrients (mmol N m⁻³)', fontsize=11)
ax2.legend(fontsize=10)
ax2.grid(True, linestyle='--', alpha=0.4)

# 3. Combined
ax3.plot(t, P, color='blue', linewidth=2.5, label='P(t) — Phytoplankton')
ax3.plot(t, N, color='green', linewidth=2.5, label='N(t) — Nutrients')
ax3.axhline(y=total_PN, color='black', linewidth=1.5, linestyle='--', label=f'P₀ + N₀ = {total_PN}')
ax3.set_title('Phytoplankton & Nutrients', fontsize=13, fontweight='bold')
ax3.set_xlabel('Time (days)', fontsize=11)
ax3.set_ylabel('Concentration (mmol N m⁻³)', fontsize=11)
ax3.legend(fontsize=10)
ax3.grid(True, linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```

#### Answer the following questions (adding plots where needed):
#### 1. Why are the curves identical in shape?
The total amount of mass must be conserved so the final amount of phytoplankton + the final amount of nutrients must be the same as the initial amount of phytoplankton + the initial amount of nutrients. The value of this P + N is 5.1 N/m3 which is indictated by the horizontal line in Figure 4. This means that every time a unit of nutrient is consumed, the same unit amount is converted into phytoplankton biomass. This means that the curves are "mirror images" of eachother. The rate of nutrient consumption will be the same as the rate of phytoplankton growth.

#### 2. What happens when you change the value of the uptake parameter? Think about it as the percentage of food consumed per day. 
The uptake parameter is k. A higher value of k which indictates that more nutrients are consumed per day (at a faster rate) produces a curve where the maximum phytoplankton biomass is reached faster and the nutrients are also depleted faster (Figure 5). 

#### 3. What happens if the initial biomass of phytoplankton is close to the amount of resource available? How does the shape of the curve change?
If the initial biomass of phytoplankton is close to the amount of nutrients available then a new equlibirum of P+N is reached for each different initial phytoplankton biomass. The shape of the curve also no longer has the typical "S" shape and instead has a faster, narrower exponential growth phase and reaches the equlibrium/maximum value much quicker. The nutrients are also consumed much quicker so is also flatter for longer since there are more phytoplankton consuming that same amount of nutrients. 

```python
# Changing the Uptake Parameter
# k′ values to explore 
k_prime_values = [0.05, 0.1, 0.2, 0.5, 1.0]   # ← add or remove values here
# time axis 
t = np.linspace(0, 20, 500)
# colours
colours = ['#2196F3', '#4CAF50', '#FF9800', '#9C27B0', '#F44336']
# plot 
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle("Figure 5: Sensitivity to k′", fontsize=18, fontweight='bold')
for k_prime, colour in zip(k_prime_values, colours):
    P = ((P0 + N0) * P0) / (P0 + N0 * np.exp(-k_prime * (P0 + N0) * t))
    N = N0 + P0 - P
    ax1.plot(t, P, color=colour, linewidth=2.5, label=f"k′ = {k_prime}")
    ax2.plot(t, N, color=colour, linewidth=2.5, label=f"k′ = {k_prime}")

# horizontal total PN line
ax1.axhline(y=total_PN, color='#333333', linewidth=1.5, linestyle='--', label=f'P₀ + N₀ = {total_PN}')
ax2.axhline(y=total_PN, color='#333333', linewidth=1.5, linestyle='--', label=f'P₀ + N₀ = {total_PN}')
# labels 
ax1.set_title('Phytoplankton Biomass', fontsize=13, fontweight='bold')
ax1.set_xlabel('Time (days)', fontsize=11)
ax1.set_ylabel('Biomass (mmol N m⁻³)', fontsize=11)
ax1.legend(fontsize=10)
ax1.grid(True, linestyle='--', alpha=0.4)

ax2.set_title('Nutrient Concentration', fontsize=13, fontweight='bold')
ax2.set_xlabel('Time (days)', fontsize=11)
ax2.set_ylabel('Nutrients (mmol N m⁻³)', fontsize=11)
ax2.legend(fontsize=10)
ax2.grid(True, linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```
